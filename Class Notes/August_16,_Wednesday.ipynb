```python
 a = int(input("Enter first number"))
 b = int(input("Enter second number"))

 smallest_number = min(a,b)

 for k in range(smallest_number,-1,-1):
  if a % k == 0 and b % k == 0:
    print("The GCD is ", k)
    break
```
